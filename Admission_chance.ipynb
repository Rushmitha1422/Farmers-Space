**Admission Chance Prediction**

Dataset Information:

Target Variable (y):

Chance of Admit

Seven features (X) are taken into consideration to predict chance of admit:

1. GRE Score
2. TOEFL Score
3. University Rating
4. SOP
5. LOR
6. CGPA               
7. Research

```python
#step 1: import library
import pandas as pd
```

```python
#step 2: import data
admn = pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/refs/heads/main/Admission%20Chance.csv')
```

```python
admn.head()
```

```python
admn.info()
```

```python
admn.describe()
```

```python
#step 3: define y and x
admn.columns
```

```python
y = admn['Chance of Admit ']
x = admn.drop(['Serial No','Chance of Admit '],axis=1)
```

```python
#step 4: split Data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=2529)
```

```python
#step 5 : select a model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
#train model
model.fit(x_train,y_train)
model.intercept_
```

```python
model.coef_
```

```python
#step 7 : prediction
y_pred = model.predict(x_test)
y_pred
```

```python
#step 8 : accuracy
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(y_test,y_pred)
```
